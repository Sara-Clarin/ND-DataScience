## Apriori algorithm

```python
# generate candidates
from collections import defaultdict
import pandas
from itertools import combinations

def generate_candidates( k , setlist ):
    Ck = []
    print(f'Generating candidates C{k+1}')
    
    # step one: self-joining with the same length(k - 1) prefix
    fk_minus1 = sorted(setlist)
    prefix_dict = defaultdict( set)
    
    if k == 1:
        Ck = [ list('a' 'b'), list('a''c'), list('a''d'), list('a''e'), list('b''c'), list('b''d'), list('b''e'), list('c' 'd'), list('c''e'), list('d''e')]
        return Ck
        
    for item in fk_minus1:       # dictionary- prefix : [suffixes]
        key = "".join(item[0: k-1] )  # 1st - k-1st elements
        value = item[ -1]             # the last element
        prefix_dict[key].add(value) 
    
    
    prefix_dict_srtd = defaultdict( list)
    for key,subset in prefix_dict.items():
        sublist = sorted(list( subset) )
        prefix_dict_srtd[key] = sublist
        
       
    # STEP 1 JOIN: pair every letter in sublist with everyother
    for key, sublist in prefix_dict_srtd.items():
        suffixes = "".join(sublist)
        combos = combinations( suffixes, 2 )
        for suffix in combos:
            Ck.append( key + "".join(suffix))
    
    print(f'Unpruned candidates: {Ck}')
     
    #STEP 2: PRUNING (check that all subsets of Ck are in Fk)
    Ck_1 = []
    for itemset_cand in Ck:   
        for k_item in fk_minus1:
            if check_membership( itemset_cand[1:], k_item):
                Ck_1.append( itemset_cand)
    
    print(f'Pruned candidates: {Ck_1}')    
    
    return Ck_1
```

```python
def check_membership(  subset, transaction):
    ''' subset: ordered list of letters
        transaction: row in the TDB we're checking
    '''
    A = set(subset)
    if A.issubset( set(transaction)):
        return True
    return False
    
```
