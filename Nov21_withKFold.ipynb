## 1. Data loading

```python
import pandas as pd
df = pd.read_csv("data/extracted_processed.csv")
df
```

```python
# find the most frequent words
from collections import Counter
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords

all_text = df['prcessed_text'].tolist()
Counter = Counter( all_text[0].split())
blacklist = [ 'to', 'and', 'they', 'but', 'their', 'by', 'the', 'be']

for article in all_text[0:1000]:
    
    article = article.split()
    Counter.update( article)

top_hundred = Counter.most_common( 100)
for item in top_hundred:
    if item[0] in stopwords.words('english') or item[0] in blacklist:
        top_hundred.remove(item)
        
print(f' TOP 100 WORDS (including stopwords)')
print(" ")
print(top_hundred)
```

```python
def count_freq( text, word):
    
    text = text.split()
    return text.count( word)
```

```python
# make data frames and counts out of the most frequent words

for item in top_hundred:
    #print(item)
    if item[0] not in blacklist:
        df[ item[0]] = 0

for item in top_hundred:
    if item[0] not in blacklist:
        df[ item[0] ] = df['prcessed_text'].apply(lambda cnt: count_freq(cnt, item[0]))


df


```

## 3. Classification

```python
feature_nms = df.columns
features = []
for item in df.columns:
    features.append(item)
features = features[3:]
print(f'FEATURES: {features}')
```

```python


import numpy as np

df=df.sample(frac=1).reset_index(drop=True)


label_1 = df[df['label'] == 1]     # 21417 rows
label_0 = df[df['label'] == 0]     # 23481 rows

total_rows = df.shape[0]
training_rows = int(df.shape[0] * 0.8) #35918
testing_rows = int(df.shape[0] - training_rows)

training_1_int = int(0.8 * label_1.shape[0])
training_0_int = int(0.8 * label_0.shape[0])

train_1 = label_1.iloc[:training_1_int]
train_0 = label_0.iloc[:training_0_int]
test_1 = label_1.iloc[training_1_int:]
test_0 = label_0.iloc[training_0_int:]

training = pd.concat([train_1, train_0], sort=False)
training = training.sample(frac=1).reset_index(drop=True)
test = pd.concat([test_1, test_0], sort=False)
test = test.sample(frac=1).reset_index(drop=True)

#X_train = training[feature_names]
X_train = training[ features]
y_train = training['label']

#X_test = test[ feature_names]
X_test = test[ features]
y_test = test['label']

print(test[test['label'] == 0])
```

# Evaluation: Simple testing and training

```python
import numpy as np


X_train, X_test = np.array(X_train), np.array(X_test)

print("X_train.shape", X_train.shape)
print("y_train.shape", len(y_train), "#positives (more cases):", sum(y_train))
print("X_test.shape", X_test.shape)
print("y_test.shape", len(y_test), "#positives (more cases):", sum(y_test))
```

```python
from sklearn import metrics

from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier

clf_knn = KNeighborsClassifier()
clf_id3 = tree.DecisionTreeClassifier(criterion="entropy", max_depth=2)
clf_id3_overfit = tree.DecisionTreeClassifier(criterion="entropy")
clf_cart = tree.DecisionTreeClassifier(max_depth=2)
clf_bayes = GaussianNB()
clf_rbf = SVC()
clf_forest = RandomForestClassifier()
clf_boost = AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=2), random_state=21)
clf_mlp = MLPClassifier(hidden_layer_sizes=(50,50,), max_iter=1000, tol=0.001, random_state=42)

clf_knn.fit(X_train, y_train)
clf_id3.fit(X_train, y_train)
clf_id3_overfit.fit(X_train, y_train)
clf_cart.fit(X_train, y_train)
clf_bayes.fit(X_train, y_train)
clf_rbf.fit(X_train, y_train)
clf_forest.fit(X_train, y_train)
clf_boost.fit(X_train, y_train)
clf_mlp.fit(X_train, y_train)

print("Metrics:")
print(f"- KNN: Accuracy: {metrics.accuracy_score(y_test, clf_knn.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_knn.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_knn.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_knn.predict(X_test)):0.4f}")
print(f"- ID3: Accuracy: {metrics.accuracy_score(y_test, clf_id3.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_id3.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_id3.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_id3.predict(X_test)):0.4f}")
print(f"- ID3 (overfitting): Accuracy {metrics.accuracy_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}")
print(f"- CART: Accuracy:  {metrics.accuracy_score(y_test, clf_cart.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_cart.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_cart.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_cart.predict(X_test)):0.4f}")
print(f"- Naive Bayes: Accuracy: {metrics.accuracy_score(y_test, clf_bayes.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_bayes.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_bayes.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_bayes.predict(X_test)):0.4f}")
print(f"- RBF Kernel SVC: Accuracy: {metrics.accuracy_score(y_test, clf_rbf.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_rbf.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_rbf.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_rbf.predict(X_test)):0.4f}")
print(f"- Random Forest: Accuracy: {metrics.accuracy_score(y_test, clf_forest.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_forest.predict(X_test)):0.4f}, Precison: {metrics.precision_score(y_test, clf_forest.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_forest.predict(X_test)):0.4f}")
print(f"- AdaBoost: Accuracy: {metrics.accuracy_score(y_test, clf_boost.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_boost.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_boost.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_boost.predict(X_test)):0.4f}")
print(f"- MLP: Accuracy: {metrics.accuracy_score(y_test, clf_mlp.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_mlp.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_mlp.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_mlp.predict(X_test)):0.4f}")
```

```python
feature_nms = df.columns
features = []
for item in df.columns:
    features.append(item)
features = features[3:]
print(f'FEATURES: {features}')
```

```python
# TODO: perform k-fold cross validation with k = 10
#    --> so 0.1 is for testing and 0.9 is for training

from collections import defaultdict
from sklearn.model_selection import KFold
from sklearn import metrics

from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier

# MAY NOT NEED: FOR TAKING MEAN, STDEV OF CROSS-VALIDATION
KNN_stats = defaultdict(list)
ID3_stats = defaultdict(list)
id3_overfit_stats = defaultdict(list)
CART_stats = defaultdict(list)
NaiveBayes_stats = defaultdict(list)
RBFKernel_stats = defaultdict(list) # format--> accuracy : [ a1, a2, a3...a10], {F1} : [f11, f12, f13....f110]
Forest_stats = defaultdict(list)
Adaboost_stats = defaultdict(list)
mlp_stats = defaultdict(list)

# train each model, give output

# CREATE MODELS
clf_knn = KNeighborsClassifier()
clf_id3 = tree.DecisionTreeClassifier(criterion="entropy", max_depth=2)
clf_id3_overfit = tree.DecisionTreeClassifier(criterion="entropy")
clf_cart = tree.DecisionTreeClassifier(max_depth=2)
clf_bayes = GaussianNB()
clf_rbf = SVC()
clf_forest = RandomForestClassifier()
clf_boost = AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=2), random_state=21)
clf_mlp = MLPClassifier(hidden_layer_sizes=(50,50,), max_iter=1000, tol=0.001, random_state=42)

# https://scikit-learn.org/0.15/modules/generated/sklearn.cross_validation.KFold.html
#x = np.array(     )
#y = np.array(     )
datarange = range(0, 44898 )

kf = KFold( n_splits=10, shuffle=False)  # switch to true

itr = 1
for train_index, test_index in kf.split( datarange):
    print(f'********************************************')
    print(f'iteration {itr}')
    X_train = df.iloc[train_index].loc[:, features]
    X_test = df.iloc[test_index][features]
    
    Y_train = df.iloc[train_index].loc[:,'label']
    y_test = df.loc[test_index]['label']
    
    # train the models
    clf_knn.fit(X_train, Y_train)
    clf_id3.fit(X_train, Y_train)
    clf_id3_overfit.fit(X_train, Y_train)
    clf_cart.fit(X_train, Y_train)
    clf_bayes.fit(X_train, Y_train)
    clf_rbf.fit(X_train, Y_train)
    clf_forest.fit(X_train, Y_train)
    clf_boost.fit(X_train, Y_train)
    clf_mlp.fit(X_train, Y_train)

    
    # output results, record statistics
    print(f"- KNN: Accuracy: {metrics.accuracy_score(y_test, clf_knn.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_knn.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_knn.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_knn.predict(X_test)):0.4f}")
    KNN_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_knn.predict(X_test)))
    KNN_stats['Recall'].append(metrics.recall_score(y_test, clf_knn.predict(X_test)))
    KNN_stats['Precision'].append(metrics.precision_score(y_test, clf_knn.predict(X_test)))
    KNN_stats['F1'].append(metrics.f1_score(y_test, clf_knn.predict(X_test)))

    print(f"- ID3: Accuracy: {metrics.accuracy_score(y_test, clf_id3.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_id3.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_id3.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_id3.predict(X_test)):0.4f}")
    ID3_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_id3.predict(X_test)))
    ID3_stats['Recall'].append(metrics.recall_score(y_test, clf_id3.predict(X_test)))
    ID3_stats['Precision'].append(metrics.precision_score(y_test, clf_id3.predict(X_test)))
    ID3_stats['F1'].append(metrics.f1_score(y_test, clf_id3.predict(X_test)))
    
    print(f"- ID3 (overfitting): Accuracy {metrics.accuracy_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_id3_overfit.predict(X_test)):0.4f}")
    id3_overfit_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_id3_overfit.predict(X_test)))
    id3_overfit_stats['Recall'].append(metrics.recall_score(y_test, clf_id3_overfit.predict(X_test)))
    id3_overfit_stats['Precision'].append(metrics.precision_score(y_test, clf_id3_overfit.predict(X_test)))
    id3_overfit_stats['F1'].append(metrics.f1_score(y_test, clf_id3_overfit.predict(X_test)))
    
    print(f"- CART: Accuracy:  {metrics.accuracy_score(y_test, clf_cart.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_cart.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_cart.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_cart.predict(X_test)):0.4f}")
    CART_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_cart.predict(X_test)))
    CART_stats['Recall'].append(metrics.recall_score(y_test, clf_cart.predict(X_test)))
    CART_stats['Precision'].append(metrics.precision_score(y_test, clf_cart.predict(X_test)))
    CART_stats['F1'].append(metrics.f1_score(y_test, clf_cart.predict(X_test)))
    
    print(f"- Naive Bayes: Accuracy: {metrics.accuracy_score(y_test, clf_bayes.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_bayes.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_bayes.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_bayes.predict(X_test)):0.4f}")
    NaiveBayes_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_bayes.predict(X_test)))
    NaiveBayes_stats['Recall'].append(metrics.recall_score(y_test, clf_bayes.predict(X_test)))
    NaiveBayes_stats['Precision'].append(metrics.precision_score(y_test, clf_bayes.predict(X_test)))
    NaiveBayes_stats['F1'].append(metrics.f1_score(y_test, clf_bayes.predict(X_test)))
    
    print(f"- RBF Kernel SVC: Accuracy: {metrics.accuracy_score(y_test, clf_rbf.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_rbf.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_rbf.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_rbf.predict(X_test)):0.4f}")
    RBFKernel_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_rbf.predict(X_test)))
    RBFKernel_stats['Recall'].append(metrics.recall_score(y_test, clf_rbf.predict(X_test)))
    RBFKernel_stats['Precision'].append(metrics.precision_score(y_test, clf_rbf.predict(X_test)))
    RBFKernel_stats['F1'].append(metrics.f1_score(y_test, clf_rbf.predict(X_test)))
    
    print(f"- Random Forest: Accuracy: {metrics.accuracy_score(y_test, clf_forest.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_forest.predict(X_test)):0.4f}, Precison: {metrics.precision_score(y_test, clf_forest.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_forest.predict(X_test)):0.4f}")
    Forest_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_forest.predict(X_test)))
    Forest_stats['Recall'].append(metrics.recall_score(y_test, clf_forest.predict(X_test)))
    Forest_stats['Precision'].append(metrics.precision_score(y_test, clf_forest.predict(X_test)))
    Forest_stats['F1'].append(metrics.f1_score(y_test, clf_forest.predict(X_test)))
    
    print(f"- AdaBoost: Accuracy: {metrics.accuracy_score(y_test, clf_boost.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_boost.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_boost.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_boost.predict(X_test)):0.4f}")
    Adaboost_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_boost.predict(X_test)))
    Adaboost_stats['Recall'].append(metrics.recall_score(y_test, clf_boost.predict(X_test)))
    Adaboost_stats['Precision'].append(metrics.precision_score(y_test, clf_boost.predict(X_test)))
    Adaboost_stats['F1'].append(metrics.f1_score(y_test, clf_boost.predict(X_test)))
    
    print(f"- MLP: Accuracy: {metrics.accuracy_score(y_test, clf_mlp.predict(X_test)):0.4f}, Recall: {metrics.recall_score(y_test, clf_mlp.predict(X_test)):0.4f}, Precision: {metrics.precision_score(y_test, clf_mlp.predict(X_test)):0.4f}, F1: {metrics.f1_score(y_test, clf_mlp.predict(X_test)):0.4f}")
    mlp_stats['Accuracy'].append(metrics.accuracy_score(y_test, clf_mlp.predict(X_test)))
    mlp_stats['Recall'].append(metrics.recall_score(y_test, clf_mlp.predict(X_test)))
    mlp_stats['Precision'].append(metrics.precision_score(y_test, clf_mlp.predict(X_test)))
    mlp_stats['F1'].append(metrics.f1_score(y_test, clf_mlp.predict(X_test)))        
                
    itr += 1
    

    
    

```
